# Day 2: Correlation Analysis Using Heatmaps

## Problem Description

Compute and visualize the correlation matrix between delay durations and numerical features using the Airline Delay dataset.

Dataset source: https://www.kaggle.com/datasets/patrickzel/flight-delay-and-cancellation-dataset-2019-2023

```python
import sys
import subprocess

for module in ["pandas", "seaborn", "matplotlib"]:
    try:
        __import__(module)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", module])
```

```python

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

url = 'mathematical_datasets/flights_sample_3m.csv'
data = pd.read_csv(url)

data.columns
```

```python
# Filter relevant numerical columns
numerical_data = data[['DEP_DELAY', 'TAXI_OUT', 'TAXI_IN', 'DISTANCE']]

# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()

# Visualize the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
